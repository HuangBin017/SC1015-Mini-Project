```python
numData = pd.DataFrame(phishdata[["total_of_www", "longest_word_path", "phish_hints", "nb_hyperlinks", "ratio_intHyperlinks", "ratio_extHyperlinks", "domain_age", "web_traffic", "google_index",  "page_rank"]])
```

```python
numData.describe()
```

```python
f, axes = plt.subplots(10, 3, figsize=(24, 24))

count = 0

for var in numData:
    sb.boxplot(data = numData[var], orient = "h", ax = axes[count,0])
    sb.histplot(data = numData[var], ax = axes[count,1])
    sb.violinplot(data = numData[var], orient = "h", ax = axes[count,2])
    count += 1
```

```python
print(numData.corr())

# Heatmap of the Correlation Matrix
f = plt.figure(figsize=(10, 10))
sb.heatmap(numData.corr(), vmin = -1, vmax = 1, linewidths = 1,
           annot = True, fmt = ".2f", annot_kws = {"size": 18}, cmap = "RdBu")
```

```python
sb.pairplot(data = numData)
```
